```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
path_dir = 'C:/Users/patel/Documents/103- Analytic tools & Decision Making/Project/merge'
for merge in os.listdir(path_dir):
    print(merge)
```

```python
all_files = glob.glob(path_dir + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, low_memory=False)
    li.append(df)
    
FinalData = pd.concat(li, axis=0, ignore_index=True)
```
